```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Download Tesla stock data
tesla = yf.Ticker("TSLA")
```

```python
# Get historical market data (last 5 years)
hist = tesla.history(period="5y")
```

```python
# Display the first few rows of the data
print("Tesla Stock Data :")
print(hist.head())
```

```python
# Basic information about the dataframe
print("\nData Shape:", hist.shape)
print("\nColumn Names:", list(hist.columns))
```

```python
# Calculate daily returns
hist['Daily_Return'] = hist['Close'].pct_change() * 100
```

```python
# Plot the closing price history
plt.figure(figsize=(14, 7))
plt.subplot(2, 1, 1)
plt.plot(hist.index, hist['Close'], 'b-', linewidth=2)
plt.title('Tesla (TSLA) Stock Price - Last 5 Years')
plt.ylabel('Price ($)')
plt.grid(True)
```

```python
# Plot the trading volume
plt.subplot(2, 1, 2)
plt.bar(hist.index, hist['Volume'], color='g', alpha=0.5)
plt.title('Tesla (TSLA) Trading Volume')
plt.ylabel('Volume')
plt.grid(True)
plt.tight_layout()
```

```python
# Calculate some basic statistics
avg_daily_return = hist['Daily_Return'].mean()
volatility = hist['Daily_Return'].std()
max_price = hist['Close'].max()
min_price = hist['Close'].min()
current_price = hist['Close'][-1]

print(f"\nSummary Statistics:")
print(f"Current Price: ${current_price:.2f}")
print(f"Average Daily Return: {avg_daily_return:.2f}%")
print(f"Volatility (Daily): {volatility:.2f}%")
print(f"Maximum Price: ${max_price:.2f}")
print(f"Minimum Price: ${min_price:.2f}")
```

```python
hist['Close'].plot(title="Tesla (TSLA) Closing Price", figsize=(10, 5))
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Web scraping to extract Tesla revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
```

```python
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Find the revenue table
tables = soup.find_all('table')
for table in tables:
    if 'Tesla Quarterly Revenue' in table.text:
        tesla_revenue = pd.read_html(str(table))[0]
        break
```

```python
# Clean the DataFrame
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != '']
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].str.replace('[\$,]', '', regex=True).astype(float)
```

```python
# Display the last five rows of the tesla_revenue dataframe
tesla_revenue.tail()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Download Tesla stock data
tesla_data = yf.download('TSLA', start='2019-01-01', end='2024-12-31')
tesla_data.reset_index(inplace=True)
```

```python
def make_graph(stock_data, stock, title):
    plt.figure(figsize=(14, 6))
    plt.plot(stock_data['Date'], stock_data['Close'], label=f"{stock} Close Price")
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
make_graph(tesla_data, stock='TSLA', title='Tesla Stock Price Over Time')
```

```python
# Download historical GameStop stock data
gme_data = yf.download('GME', start='2019-01-01', end='2024-12-31')
gme_data.reset_index(inplace=True)
```

```python
def make_graph(stock_data, stock, title):
    plt.figure(figsize=(14, 6))
    plt.plot(stock_data['Date'], stock_data['Close'], label=f"{stock} Close Price", color='purple')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
make_graph(gme_data, stock='GME', title='GameStop Stock Price Over Time')
```
